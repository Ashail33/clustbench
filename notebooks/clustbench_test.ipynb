```
# Install the clustbench package (if not installed). You may need to run this cell.
# !pip install --quiet git+https://github.com/Ashail33/clustbench.git

# Import the package
from clustbench.datasets import gen_blobs, DataSpec
from clustbench.algorithms.kmeans import Kmeans
from clustbench.metrics import bundle_scores
```

```
# Generate a synthetic dataset
spec = DataSpec(n_samples=500, n_features=8, centers=5, compactness=0.5, seed=42)
X, y_true = gen_blobs(spec)

# Run KMeans
algo = Kmeans(n_init=10, max_iter=200, random_state=42)
res = algo.fit_predict(X, k=spec.centers)
labels = res.labels

# Compute metrics
metrics = bundle_scores(X, labels, y_true=y_true)

print('ARI:', metrics['ari'])
print('NMI:', metrics['nmi'])
print('Silhouette:', metrics['silhouette'])
```
